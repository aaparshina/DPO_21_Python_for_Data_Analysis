```python
text = "12 ноября 2011 года произошло удивительное событие. А 13 ноября 2012 - еще удивительнее. Даже не будем \
      говорить, что произошло 2 декабря 2011 года и 25 декабря 2012 года."
print(re.findall(r"\d{4}", text)) # 4 цифры подряд
```

```python
text2 = "По имеющимся данным, в Екатеринбургской губернии на май 1916 года было занято 50611 военнопленных,\
      из них 34194 на фабричных и заводских работах, 5731 на «казённых», 5060 на сельскохозяйственных,\
      4145 на железнодорожных, 913 на городских и земских, 568 на прочих."

print(re.findall(r"[12]\d{3}", text2)) 
```

```python
text = "12 ноября 2011 года произошло удивительное событие. А 13 ноября 2012 - еще удивительнее. Даже не будем \
      говорить, что произошло 2 декабря 2011 года и 25 декабря 2012 года."

print(re.findall("удивительн..", text))
```

```python
#Число, пробел, набор букв, пробел, четыре цифры
print(re.findall("\d{1,2} [а-я]+ \d{4}", text)) # осталось прочитать регулярку по слогам :)
print(re.findall("\d{1,2}\s[а-я]+\s\d{4}", text)) 
```

```python
data = '20.05.1963, 55, 12.12.2000, 17, 15/15/1111'
print(re.findall(r"\d+\.\d+.\d{4}", data)) 
```

## Задача 2

Извлечь все слова, начинающиеся на гласную

    string = 'Десять негритят отправились обедать, Один поперхнулся, их осталось девять.'

```python
string = 'Десять негритят отправились обедать, Один поперхнулся, их осталось девять.'
print(re.findall(r'\w+', string))
print(re.findall(r'[аяоеуюыиэеАЯОЕУЮЫИЭ]\w+', string))
```

```python
print(re.findall(r'\b[аяоеуюыиэеАЯОЕУЮЫИЭ]\w+', string))
```
