## Задача 3 (2 балла) - Проверить номера автомобилей

В России применяются регистрационные знаки нескольких видов. Общего в них то, что они состоят из цифр и букв. Причём используются только 12 букв кириллицы, имеющие графические аналоги в латинском алфавите — А, В, Е, К, М, Н, О, Р, С, Т, У и Х.

У частных легковых автомобилях номера — это буква, три цифры, две буквы, затем две или три цифры с кодом региона. У такси — две буквы, три цифры, затем две или три цифры с кодом региона. Есть также и другие виды, но в этой задаче они не понадобятся.

Напишите код, который с помощью регулярных выражений будет проверять номер автомобиля и печатать "тип" автомобиля -- Private или Taxi. Если проверить номер не получилось, то программа печатает Fail. Работайте со списком `numbers`. В списке `answers` даны верные ответы, которые должен выдавать ваш код (нигде в коде этот список использовать не надо). 

```python
numbers = ['С227НА777','КУ22777','Т22В7477','М227К19У9','С227НА8777']
# Теперь правильные ответы
answers = ['Private','Taxi','Fail','Fail','Fail']
```

```python
for num in numbers:
    if re.match(r'[АВЕКМНОРСТУХ]{1}\d{3}[АВЕКМНОРСТУХ]{2}\d{2,3}', num) and len(num) in (8,9):
        print('Private')
    elif re.match(r'[АВЕКМНОРСТУХ]{2}\d{5,6}', num) and len(num) in (7,8):
        print('Taxi')
    else:
        print('Fail')
```

## Задача 4 (3 балла)

В файле `mbox.txt` есть строки формата "Date: Sat, 5 Jan 2008 09:12:18 -0500" - время, когда ушло письмо. Давайте создадим словарик, в который будем сохранять, в каком часу люди пишут письма (час от 0 до 23 - ключ, количество писем, написанных в это время, - значение).

Пусть программа выводит строки типа "В <n> часов было отправлено <m> писем".     



```python
hours = {}

with open('mbox.txt') as f:
    for line in f:
        data = re.findall(r'Date:.+(\d{2}):(\d{2}):(\d{2}).+\)', line)
        if len(data) != 0:
            
            if data[0][0] in hours:
                hours[data[0][0]] += 1
            else:
                hours[data[0][0]] = 1
            #hours[data[0][0]] = hours.get(hours[data[0][0]], 0) + 1
            
for hour in sorted(hours):
    print(f'В {hour} часов было отправлено {hours[hour]} писем')
```
