```python
import googleapiclient.discovery as api
import pandas as pd
import plotly.graph_objects as go
youtube = api.build('youtube', 'v3', developerKey = 'ВАШ КЛЮЧ')
```

### На прошлом занятии: 

```python
request = youtube.search().list(
        part="snippet",
        maxResults=50,
        q="вакцинация",
        type='video')
response = request.execute()
```

```python
df = pd.json_normalize(response['items'])
df.head()
```

```python
response['nextPageToken']
```

```python
i = 0
while 'nextPageToken' in response.keys():
    
    request = youtube.search().list(
        part="snippet",
        maxResults=50,
        q="вакцинация",
        type='video',
        pageToken=response['nextPageToken'])
    
    response = request.execute()
    
    i += 1
    print(f'Итерация {i}')
    
    df_add = pd.json_normalize(response['items'])
    df = pd.concat([df, df_add])
df
```

```python
df.index = range(1, len(df)+1)
df = df.drop_duplicates('id.videoId')
df.shape
```

### Новое. Метод ``.search().list()``, параметр `order`

[Документация к методу](https://developers.google.com/youtube/v3/docs/search/list?hl=ru)

```python
order = ['date', 'rating'] 
```

```python
i = 1
for o in order:
  request = youtube.search().list(
    part = 'snippet',
    maxResults = 50,
    q = 'вакцинация',
    type = 'video',
    order = o)
  response = request.execute()

  print(f'Итерация {i}, order {o}, items {len(response["items"])}')
  i += 1

  df_add = pd.json_normalize(response['items'])
  df = pd.concat([df, df_add])

  while 'nextPageToken' in response.keys():
    request = youtube.search().list(
    part = 'snippet',
    maxResults = 50,
    q = 'вакцинация',
    type = 'video',
    order = o,
    pageToken = response['nextPageToken'])
    
    response = request.execute()

    print(f'Итерация {i}, order {o}, items {len(response["items"])}')
    i += 1

    df_add = pd.json_normalize(response['items'])
    df = pd.concat([df, df_add])

```

### Новое. Метод ``videoCategories().list()``

[Документация к методу](https://developers.google.com/youtube/v3/docs/videoCategories/list?hl=ru)

```python
data_video['snippet.categoryId'].unique()
```

```python
request = youtube.videoCategories().list(
    part = 'snippet',
    id = list(data_video['snippet.categoryId'].unique())

)
response = request.execute()
```

```python
dat_cat = pd.json_normalize(response['items'])
dat_cat = dat_cat.set_index('id')
dat_cat.head(2)
```

```python
for i in dat_cat.index:
  data_video.loc[data_video['snippet.categoryId'] == i, 'categoryName'] = dat_cat['snippet.title'][i]
```

```python
d = dict(data_video['categoryName'].value_counts())
d
```

```python
trace = go.Bar(
    x = list(d.keys()),
    y = list(d.values())
)

fig = go.Figure(data = [trace])

fig.show()
```
